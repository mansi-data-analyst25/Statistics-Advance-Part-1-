#How They Relate?

$$

> Where:

>> ğ‘¥Ë‰ = sample mean

>> Ïƒ = population standard deviation

>> n = sample size

>> Zâˆ— = Z-score (also called critical value) based on confidence level

#Z-scores for Common Confidence Levels:

| Confidence Level | Z-score ($Z^*$) |
| ---------------- | --------------- |
| 90%              | 1.645           |

These Z-scores mark the cutoff points that capture the middle portion of the distribution.

# Q 21: What are the assumptions for applying the Central Limit Theorem?

ANS: The Central Limit Theorem (CLT) is one of the most powerful results in statistics. It states that the sampling distribution of the sample mean becomes approximately normal as the sample size increases, regardless of the population's distribution â€” under certain assumptions.

#Key Assumptions of the Central Limit Theorem:

| Assumption                      | Description                                                                                                                                                                                                                                                                                                                             |
| ------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1. Random Sampling**          | The sample data must be **randomly and independently selected** from the population.                                                                                                                                                                                                                                                    |
| **2. Independent Observations** | Each observation in the sample must be **independent** of the others.                                                                                                                                                                                                                                                                   |
| **3. Identically Distributed**  | All observations must come from the **same population** (same distribution, same parameters).                                                                                                                                                                                                                                           |
| **4. Sample Size**              | The sample size $n$ should be **"sufficiently large"**:<br> - **$n \geq 30$** is a common rule of thumb if the population distribution is unknown or skewed.<br> - If the population is **normal**, even small $n$ is acceptable.<br> - For **highly skewed or non-normal populations**, larger samples (e.g., $n > 50$) may be needed. |

#When Assumptions Are Violated?

> Non-random samples can bias results.

> Dependent observations (like time series data) can distort the distribution.

> Different distributions in the data invalidate the theoretical basis.

> Too small a sample may not allow the CLT to "kick in."

#Why These Assumptions Matter?

> They ensure the sampling distribution of the mean will approach a normal distribution, which allows for:

>> Using Z-scores

>> Building confidence intervals

>> Performing hypothesis testing using the normal distribution

# Q 22: What is the concept of expected value in a probability distribution?

ANS: The expected value (often denoted as E(X) or Î¼) is the long-run average or mean of a random variable in a probability distribution. It represents the value you expect to get on average if an experiment is repeated many times.

#For Discrete Random Variables:

If \( X \) is a discrete random variable, and it takes values \( x_1, x_2, \ldots, x_n \)  
with corresponding probabilities \( P(x_1), P(x_2), \ldots, P(x_n) \), then:

>> $$
E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
$$

Multiply each outcome by its probability and add them up.

# Example:

> A fair 6-sided die has outcomes: 1 through 6

> Each outcome has probability 1/6

>> $$
E(X) = 1 \cdot \frac{1}{6} + 2 \cdot \frac{1}{6} + \cdots + 6 \cdot \frac{1}{6} = \frac{21}{6} = 3.5
$$

On average, a die roll gives 3.5 (even though it's not an actual possible result).

#For Continuous Random Variables:

If ğ‘‹ is a continuous random variable with a probability density function (PDF) ğ‘“(ğ‘¥), then:

>> $$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

This is the weighted average of all possible values of ğ‘‹, where the weights are given by the PDF.

#Intuition:

The expected value tells you:

> What you expect "on average" over many trials

> A measure of the center of the distribution

> Often used in decision-making, economics, and game theory

# Q 23: How does a probability distribution relate to the expected outcome of a random variable?

ANS: The probability distribution of a random variable describes how likely each possible outcome is. The expected outcome, or expected value, is the weighted average of those outcomes â€” using the probabilities as weights.

# The Connection Explained:

| Concept                      | Description                                                                                  |
| ---------------------------- | -------------------------------------------------------------------------------------------- |
| **Random Variable (X)**      | A variable representing outcomes of a random experiment (like die rolls, test scores, etc.). |
| **Probability Distribution** | Describes the **likelihood** of each possible value of the random variable.                  |
| **Expected Value $E(X)$**    | The **average outcome** you'd expect **over many repetitions** of the experiment.            |

The expected value is calculated directly from the probability distribution.

#Discrete Example:

Suppose a random variable ğ‘‹ has the following distribution:

| X (Outcome) | P(X) (Probability) |
| ----------- | ------------------ |

To find the expected value:

> E(X)=1â‹…0.2+2â‹…0.5+3â‹…0.3=0.2+1.0+0.9=2.1

This means that over the long run, the average outcome will be 2.1.

#Continuous Example:

If a random variable ğ‘‹ has a probability density function (PDF) ğ‘“(ğ‘¥), the expected value is:

>> $$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

This is the continuous version of â€œweighted average,â€ where the weights are infinitesimally small probabilities across the range of ğ‘¥.

# Q 3: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    Calculates the PMF of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1)
        p (float): Probability of success (0 â‰¤ p â‰¤ 1)

    Returns:
        float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)


print(bernoulli_pmf(1, 0.7))
print(bernoulli_pmf(0, 0.7))
```

# Q 18: Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a data point.

    Parameters:
        x (float): The data point
        mean (float): Mean of the distribution
        std_dev (float): Standard deviation of the distribution

    Returns:
        float: Z-score
    """
    z = (x - mean) / std_dev
    return z

def plot_standard_normal_with_z(z):
    """
    Plot the standard normal distribution and mark the given Z-score.

    Parameters:
        z (float): Z-score to highlight
    """
    x = np.linspace(-4, 4, 500)
    pdf = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, label='Standard Normal Distribution', color='navy')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='skyblue', alpha=0.6, label=f'P(Z â‰¤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title("Standard Normal Distribution with Z-score")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage
x = 85         # data point
mean = 75      # sample/population mean
std_dev = 5    # standard deviation

z = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x}: {z:.2f}")

plot_standard_normal_with_z(z)
```

# Q 19: Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Step 1: Input the data
sample_mean = 980           # sample mean (xÌ„)
population_mean = 1000      # claimed population mean (Î¼â‚€)
sample_std = 50             # sample standard deviation (s)
n = 40                      # sample size
alpha = 0.05                # significance level

# Step 2: Calculate the Z-statistic
standard_error = sample_std / np.sqrt(n)
z_stat = (sample_mean - population_mean) / standard_error

# Step 3: Calculate the p-value (one-tailed, left-side)
p_value = norm.cdf(z_stat)

# Step 4: Make decision
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Evidence suggests the bulbs last less than 1000 hours.")
else:
    print("Fail to reject the null hypothesis: No strong evidence bulbs last less than 1000 hours.")
```

# Q 20: Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# Step 1: Sample dataset (e.g., product weights in grams)
data = [50.2, 49.7, 51.0, 50.5, 49.9, 50.1, 50.4, 50.3, 49.8, 50.0]

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # use ddof=1 for sample standard deviation
n = len(data)

# Step 3: Set confidence level (e.g., 95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 4: Calculate Z critical value for 95% CI
z_critical = norm.ppf(1 - alpha/2)

# Step 5: Calculate standard error and margin of error
standard_error = sample_std / np.sqrt(n)
margin_of_error = z_critical * standard_error

# Step 6: Compute confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_val = 4  # average number of events (Î»)
x_values = np.arange(0, 15)  # range of values for x

# Calculate the CDF
cdf_values = poisson.cdf(x_values, mu=lambda_val)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, where='post', color='darkgreen', label='CDF')
plt.title(f"Cumulative Distribution Function of Poisson(Î» = {lambda_val})")
plt.xlabel("x (number of events)")
plt.ylabel("P(X â‰¤ x)")
plt.xticks(x_values)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

# Example interpretation
x_check = 6
prob = poisson.cdf(x_check, mu=lambda_val)
print(f"P(X â‰¤ {x_check}) for Poisson(Î»={lambda_val}) = {prob:.4f}")
```

# Q 25:  Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two datasets
np.random.seed(42)  # for reproducibility

data1 = np.random.normal(loc=50, scale=5, size=100)   # smaller std deviation
data2 = np.random.normal(loc=50, scale=15, size=100)  # larger std deviation

# Step 2: Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Step 3: Visualize the distributions
plt.figure(figsize=(10, 5))
plt.hist(data1, bins=15, alpha=0.6, label=f'Dataset 1 (std = {std1:.2f})', color='skyblue', edgecolor='black')
plt.hist(data2, bins=15, alpha=0.6, label=f'Dataset 2 (std = {std2:.2f})', color='salmon', edgecolor='black')
plt.axvline(np.mean(data1), color='blue', linestyle='--', label='Mean of Dataset 1')
plt.axvline(np.mean(data2), color='red', linestyle='--', label='Mean of Dataset 2')

plt.title("Comparison of Two Datasets by Spread (Standard Deviation)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Q 26:  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 100            # mean
sigma = 15          # standard deviation
n = 1000            # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate Interquartile Range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Display Results
print(f"Range: {data_range:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

# Q 27:  Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset (e.g., normal but not standardized)
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)  # mean=100, std=20

# Step 2: Apply Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_normalized = (data - mean) / std

# Step 3: Plot original vs normalized data
plt.figure(figsize=(12, 5))

# Original distribution
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)

# Normalized distribution
plt.subplot(1, 2, 2)
plt.hist(z_normalized, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title("Z-score Normalized Data Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Step 4: Confirm mean and std
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(z_normalized):.2f}, Std Dev: {np.std(z_normalized):.2f}")
```

# Q 28: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
        mean (float): Mean of the distribution
        std (float): Standard deviation
        size (int): Number of samples

    Returns:
        dict: Skewness and kurtosis of the dataset
    """
    # Generate data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # use fisher=False to get Pearson's definition (normal â†’ 3)

    # Print results
    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f}")

    return {"skewness": skew_val, "kurtosis": kurt_val}
analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000)
```
